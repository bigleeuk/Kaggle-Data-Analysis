```python
friend_dict_list=[
    {"name":"jone", "midterm":9,'final':90},
    {"name":"kane", "midterm":10,'final':95},
    {"name":"pane", "midterm":11,'final':100}
]
```

```python
df = pd.DataFrame(friend_dict_list,columns =['name','midterm','final'])
```

```python
df
```

```python
df['total']=df['midterm']+df['final']
```

```python
df
```

```python
df['avg']= df['total']/2
```

```python
df
```

```python
grade=[]
for row in df['avg']:
    if row>50:
        grade.append('A')
    else:
        grade.append('B')
```

```python
grade
```

```python
df['grade']=grade
```

```python
df
```

```python
def passOrFail(row):
    if row == 'B':
        return "pass"
    else:
        return "fail"
```

```python
df.grade = df.grade.apply(passOrFail)
```

```python
 df
```
