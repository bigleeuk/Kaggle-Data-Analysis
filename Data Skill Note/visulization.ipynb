먼저 시각화에서는 보통 matplotlib을 주로 사용하며 그중 seaborn을 자주 사용합니다.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1,df2=[1,2,3,],["seoul","paris","rome"]
```

```python
plt.plot(df2,df1)
```

왼쪽이 x좌표 오른쪽이 y좌표에 해당하는 리스트이며 각각 매칭이되어 그래프가 그려지는 것을 볼 수 있습니다. 만약 축에 이름을 설정하고 싶다면 다음과같이 하시면됩니다.

```python
plt.xlabel("city")
plt.ylabel("pollution rate")
plt.plot(df2,df1)
```

이번에는 하나 더 그래프를 추가하고 그래프의 이름을 설정해보도록 하겠습니다.

```python
plt.xlabel("city")
plt.ylabel("pollution rate")
plt.plot(df2,df1)
df3,df4=[3,4,5],["hongkong","seattle","seoul"]
plt.plot(df4,df3)
plt.legend(['2017','2018'])
plt.title("each cities pollution rate")
```

그래프를 보니 두번째 그래프가 비정상처럼 보일수도 있지만 지극히 정상적으로 출력되었습니다. 이는 plot이 선을 연결하여 보여주는 함수이기 때문입니다. 그러면 각각의 수치에 따른 변화율을 보여주고자 할때는 barplot이 더 편리하고 한 눈에 알아볼 수 있을것입니다.

```python
plt.xlabel("city")
plt.ylabel("pollution rate")
plt.bar(df2,df1,width=0.8,color="red")
df3,df4=[3,4,5],["hongkong","seattle","seoul"]
plt.bar(df4,df3,width=0.4,color="blue")
plt.legend(['2017','2018'])
plt.title("each cities pollution rate")
```
