먼저 시각화에서는 보통 matplotlib을 주로 사용하며 그중 seaborn을 자주 사용합니다.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1,df2=[1,2,3,],["seoul","paris","rome"]
```

```python
plt.plot(df2,df1)
```

왼쪽이 x좌표 오른쪽이 y좌표에 해당하는 리스트이며 각각 매칭이되어 그래프가 그려지는 것을 볼 수 있습니다. 만약 축에 이름을 설정하고 싶다면 다음과같이 하시면됩니다.

```python
plt.xlabel("city")
plt.ylabel("pollution rate")
plt.plot(df2,df1)
```

이번에는 하나 더 그래프를 추가하고 그래프의 이름을 설정해보도록 하겠습니다.

```python
plt.xlabel("city")
plt.ylabel("pollution rate")
plt.plot(df2,df1)
df3,df4=[3,4,5],["hongkong","seattle","seoul"]
plt.plot(df4,df3)
plt.legend(['2017','2018'])
plt.title("each cities pollution rate")
```

그래프를 보니 두번째 그래프가 비정상처럼 보일수도 있지만 지극히 정상적으로 출력되었습니다. 이는 plot이 선을 연결하여 보여주는 함수이기 때문입니다. 그러면 각각의 수치에 따른 변화율을 보여주고자 할때는 barplot이 더 편리하고 한 눈에 알아볼 수 있을것입니다.

```python
plt.xlabel("city")
plt.ylabel("pollution rate")
plt.bar(df2,df1,width=0.8,color="red")
df3,df4=[3,4,5],["hongkong","seattle","seoul"]
plt.bar(df4,df3,width=0.4,color="blue")
plt.legend(['2017','2018'])
plt.title("each cities pollution rate")
```

다음은 Scatter 시각화에 대해 설명드리겠습니다. 말그대로 점을 찍는 것입니다. 설명보다 코드로 보는게 훨신 이해가 빠르겠죠? ^^

```python
plt.scatter(df3,df4)
```

위처럼 x 변수에 대한 y값으로 점들이 찍히게 됩니다. 파이 그래프는 어떻게 될까요?

```python
plt.pie(df3)
```

그러면 이번에는 여러개의 그래프를 동시에 그려보도록 하겠습니다. 먼저 코드를 보겠습니다.

```python
fig,ax=plt.subplots(1,3)
ax[0].bar(df3,df4)
ax[1].scatter(df3,df4)
ax[2].pie(df3)
```

한 번에 여러개의 그래프를 그리려면 plt.subplots(행, 열) 를 이용하면 됩니다 plt.subplots(행, 열) 은 2개의 변수를 반환하는데,
하나는 전체 액자 figure에 대한 변수고, 다른 하나는 액자 내 여러개의 액자에 대한 리스트 axs입니다.~

각 액자에 그래프를 그리려면 다음과 같이 plt.plot() 이 아니라, ax[i].plot() 을 사용하면 된다. ax 는 리스트로, ax[0], ax[1] .. 에 각 액자를 하나씩 담고있습니다.

이번에는 그래프의 크기, 색상 등에 대해 알아보도록 하겠습니다.

```python
fig, ax = plt.subplots(1, 4, figsize=(20, 5))
ax[0].plot([1,2,3], [1,2,3], color='C0') # matplotlib 에 정의된 색상을 써도 된다. 
ax[1].plot([1,2,3], [1,2,3], color='r') # 'red' 대신 'r'만 써도 된다. 
ax[2].plot([1,2,3], [1,2,3], color='#aa1f3c') # 16진수 표기법도 가능하다. 
ax[3].plot([1,2,3], [1,2,3], color=(0.5, 0.2, 0.1)) # 0~1로 정규화된 rgb 값도 가능하다. 
```

시각화 코드들을 보다보면, plt.figure() 을 사용하지 않고, 주로 plt.subplots()을 사용하는 경우가 많은데, 이 때 ax를 이용하여 설정할 수 있습니다.

```python
fig, ax = plt.subplots(1, 1) 
ax.plot([1,2,3], [1,2,3]) 
ax.set_title("제목") 
ax.set_xlabel("x축 이름") 
ax.set_ylabel("y축 이름") 
plt.show()
```

위 오류는 폰트 때문에 생긴문제라 넘기셔도 됩니다 정상작동된겁니다~

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 5)) 
# 0번 그래프 
ax[0].plot([1,2,3], [1,2,3], color='C0') 
ax[0].set_title("0번 제목") 
ax[0].set_xlabel("0번 x축 이름") 
ax[0].set_ylabel("0번 y축 이름") 
# 1번 그래프 
ax[1].plot([1,2,3], [1,2,3], color='C1') 
ax[1].set_title("1번 제목") 
ax[1].set_xlabel("1번 x축 이름") 
ax[1].set_ylabel("1번 y축 이름") 
plt.show()
```

x축, y축 단위와 범위 바꾸기를 알아보도록 하겠습니다.

이전과 같이 그래프를 '그냥' 그리면, 각 데이터를 파악해서 plt.plot 이 '알아서' 이쁘게 그려줍니다. 이 과정 중에, 각 축의 단위와 범위도 자연스럽게 자동으로 설정되는데, 이를 좀 더 커스터마이징 해보자. 축 단위 바꾸기 plt.xticks([단위 리스트]) 를 사용하면 x축 단위를 수동으로 바꿀 수 있다.y축도 마찬가지

```python
plt.figure() 
plt.plot([1,2,3], [1,2,3]) 
plt.xticks([1,2,3]) # x축 단위 바꾸기 
plt.yticks([1,2,3]) # y축 단위 바꾸기 plt.show()
```

```python
plt.figure() 
plt.plot([1,2,3], [1,2,3]) 
plt.xticks([1,2,3], ['월', '화', '수']) # x축 단위 바꾸기 
plt.yticks([1,2,3], ['저', '중', '고']) # y축 단위 바꾸기 plt.show()
```
